## Data Analysis Question 1

The two most important customers of the firm may have been hacked. Verify if there are any fraudulent transactions in their history. For privacy reasons, you only know that their cardholder IDs are 2 and 18.

* Using hvPlot, create a line plot representing the time series of transactions over the course of the year for each cardholder separately. 

* Next, to better compare their patterns, create a single line plot that containins both card holders' trend data.  

* What difference do you observe between the consumption patterns? Does the difference suggest a fraudulent transaction? Explain your rationale in the markdown report.

```python
# loading data for card holder 2 and 18 from the database
# Write the query
query = """
        SELECT tr.date, tr.amount AS daily_amount, cc.cardholder_id
            FROM transaction AS tr JOIN credit_card AS cc 
            ON tr.card=cc.card 
            WHERE cc.cardholder_id IN (2, 18) ;
        """
# Create a DataFrame from the query result. HINT: Use pd.read_sql(query, engine)
df = pd.read_sql(query, engine)
df['daily_amount'] = pd.to_numeric(df['daily_amount'].str.strip('$').str.replace(',',''))
type(df['daily_amount'].values[0])
```

```python
# Plot for cardholder 2
plot_2 = df[df['cardholder_id']==2].hvplot(
    x='date',
    y='daily_amount',
    title='Cardholder 2 transactions',
    label='Cardholder 2'
)
plot_2
```

```python
# Plot for cardholder 18
plot_18 = df[df['cardholder_id']==18].hvplot(
    x='date',
    y='daily_amount',
    title='Cardholder 18 transactions',
    label='Cardholder 18'
)
plot_18
```

```python
# Combined plot for card holders 2 and 18
combined_plot = plot_2 * plot_18
combined_plot.opts(title='Cardholder 2 and 18 transactions')
```

## Data Analysis Question 2

The CEO of the biggest customer of the firm suspects that someone has used her corporate credit card without authorization in the first quarter of 2018 to pay quite expensive restaurant bills. Again, for privacy reasons, you know only that the cardholder ID in question is 25.

* Using hvPlot, create a box plot, representing the expenditure data from January 2018 to June 2018 for cardholder ID 25.

* Are there any outliers for cardholder ID 25? How many outliers are there per month?

* Do you notice any anomalies? Describe your observations and conclusions in your markdown report.

```python
# loading data of daily transactions from jan to jun 2018 for card holder 25
# Write the query
query = """
        SELECT tr.date::DATE, tr.amount AS daily_amount
            FROM transaction AS tr
            JOIN credit_card AS cc
            ON tr.card=cc.card
            WHERE cc.cardholder_id=25 AND
            tr.date BETWEEN '2018-01-01' AND '2018-06-30' ;
        """
# Create a DataFrame from the query result. HINT: Use pd.read_sql(query, engine)
df = pd.read_sql(query,engine)
df['daily_amount'] = pd.to_numeric(df['daily_amount'].str.strip('$').str.replace(',',''))
type(df['date'][0])
```

```python
# loop to change the numeric month to month names
df1 = df.copy()
for row_no in range(len(df)):
    the_date = df.iloc[row_no,0]
    month = the_date.strftime('%B')
    df1.iloc[row_no,0] = month

df_ja = df1[df1['date']=='January'].rename(columns={'daily_amount':'January'}).drop('date',axis=1)
df_fe = df1[df1['date']=='February'].rename(columns={'daily_amount':'February'}).drop('date',axis=1)
df_mr = df1[df1['date']=='March'].rename(columns={'daily_amount':'March'}).drop('date',axis=1)
df_ap = df1[df1['date']=='April'].rename(columns={'daily_amount':'April'}).drop('date',axis=1)
df_ma = df1[df1['date']=='May'].rename(columns={'daily_amount':'May'}).drop('date',axis=1)
df_jn = df1[df1['date']=='June'].rename(columns={'daily_amount':'June'}).drop('date',axis=1)
```

```python
# Creating the six box plots using hvPlot
plot_ja = df_ja.hvplot.box(label='January',hover_cols=['daily_amount']) 
plot_fe = df_fe.hvplot.box(label='February',hover_cols=['daily_amount'])
plot_mr = df_mr.hvplot.box(label='March',hover_cols=['daily_amount']) 
plot_ap = df_ap.hvplot.box(label='April',hover_cols=['daily_amount'])
plot_ma = df_ma.hvplot.box(label='May',hover_cols=['daily_amount']) 
plot_jn = df_jn.hvplot.box(label='June',hover_cols=['daily_amount'])
plot_comb = plot_ja * plot_fe * plot_mr * plot_ap * plot_ma * plot_jn
plot_comb.opts(frame_height=400)
```
